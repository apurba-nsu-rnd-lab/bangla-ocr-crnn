```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def addlabels(labels, data, xadj):
    for i in range(len(labels)):
        plt.text(i+xadj, data[i]+1, data[i], horizontalalignment='center', verticalalignment='center')

# create data
x = np.arange(6)
y1 = [03.59, 16.67, 12.63, 30.38, 01.08, 17.36]
y2 = [02.84, 11.90, 09.26, 21.52, 01.08, 10.74]
y3 = [03.00, 14.29, 08.04, 21.20, 00.81, 11.57]
labels = ['I', 'II', 'III', 'IV', 'V', 'VI']
legends = ["CRNN-Naive", "CRNN-VDS", "CRNN-ADS"]
width = 0.3

fig = plt.figure(figsize = (12, 5), dpi=300)

cmap = plt.get_cmap("Oranges")
colors = cmap([0.2, 0.4, 0.6, 0.8])
        
# plot data in grouped manner of bar type
plt.bar(x-width, y1, width, color=colors[0])
plt.bar(x, y2, width, color=colors[1])
plt.bar(x+width, y3, width, color=colors[2])
plt.xticks(x, labels)
plt.xlabel("Protocol")
plt.ylabel("Percentage of mispredicted consonant clusters")
plt.legend(legends, loc='upper left')
addlabels(labels, y1, xadj = -width)
addlabels(labels, y2, xadj = 0)
addlabels(labels, y3, xadj = width)
    
plt.show()
```
