```python
import pandas as pd
```

```python
majedul = pd.read_csv('Updated_CharactersInBanglaLanguage.csv')
```

```python
apurba_char_list = majedul['chars'].tolist()
```

```python
apurba_char_list
```

```python
our_char = pd.read_csv('character_frequency_synthetic2m.csv')
our_grapheme = pd.read_csv('grapheme_frequency_synthetic2m.csv')
```

```python
our_char_list = our_char['Character'].tolist()
our_grapheme_list = our_grapheme['Grapheme'].tolist()
```

```python
concatenated_list = apurba_char_list + our_char_list + our_grapheme_list
```

```python
new_list = set(concatenated_list)
print('length of new list:', len(new_list))
new_list = list(new_list)
new_list.sort()
new_list
```

```python
newly_added = set(new_list) - set(apurba_char_list)
```

```python
newly_added
```

```python
df = pd.DataFrame(data={"Character": new_list})
df.to_csv("./file.csv", sep=',',index=False)
```

```python
df = pd.DataFrame(data={"Character": list(newly_added)})
df.to_csv("./file.csv", sep=',',index=False)
```
